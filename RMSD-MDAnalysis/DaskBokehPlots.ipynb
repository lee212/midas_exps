This file brings together code from Dask Distributed Diagnostics
and Dask Distributed Bokeh. Along with the Dask Distributed Scheduler Plugin 
can be used to create post-mortem Dask Task Stream Bokeh plots

author: iparask

```python
from bokeh.plotting import figure, output_file,show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool
from bokeh.layouts import gridplot
import pandas as pd
from time import sleep
from DaskProfileReader import *
```

```python
file_size=[100,300]
blocksize=[16,32,48,64]
for size in file_size:
    for block in blocksize:
        for run in range(1,6):
            name = 'XTC_{}_{}_{}'.format(size,block,run)
            profile = DaskProfileReader('Data/'+name+'.txt')
            print name
            output_file('DaskBokehplots/'+name+'.html')
            profile['dt']=profile['start']-profile['start'].min()
            half = (profile['duration'][profile[profile['start']==profile['start'].min()].index])*0.5
            profile['dt'] = profile['dt']+half.values
            CDS=ColumnDataSource(profile)
            plot=figure(height=600,width=1000,x_axis_label='Time in msec',y_axis_label='Worker Proccess',title='XTC {}x {} cores run {}'.format(size,block,run))
            hover=HoverTool(point_policy="follow_mouse",
                        tooltips="""
                            <div>
                                <span style="font-size: 12px; font-weight: bold;">@name:</span>&nbsp;
                                <span style="font-size: 12px; font-weight: bold;">@action:</span>&nbsp;
                                <span style="font-size: 10px; font-family: Monaco, monospace;">@duration</span>
                                <span style="font-size: 10px;">ms</span>&nbsp;
                            </div>
                        """
                    )
            plot.add_tools(hover)
            plot.rect('dt','y',height=0.5,width='duration',color='color',alpha='alpha',source=CDS)
            show(plot)
            sleep(10)
```
