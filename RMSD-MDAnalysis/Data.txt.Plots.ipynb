```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xtcRuns=pd.DataFrame.from_csv('data.csv',index_col='Index')
```

```python
xtcRuns
```

```python
XTCmeans100=pd.DataFrame(columns=xtcRuns.columns)
XTCmeans300=pd.DataFrame(columns=xtcRuns.columns)

for i in range(0,20,5):
    XTCmeans100.loc[len(XTCmeans100)]=xtcRuns[i:i+5].mean()
    XTCmeans300.loc[len(XTCmeans300)]=xtcRuns[i+20:i+25].mean()
    
XTCmeans100 = XTCmeans100.drop(['Type','Run'],1)
XTCmeans300 = XTCmeans300.drop(['Type','Run'],1)

XTCmeans100
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1)
XTCmeans100.plot(x='Block',y='t_comp_avg',ax=axes,label='XTC100',style='-*')
XTCmeans300.plot(x='Block',y='t_comp_avg',ax=axes,label='XTC300',style='-o')
axes.set_xticks([16,32,48,64])
axes.set_xlabel('Blocks',fontsize=14)
axes.set_ylabel('Time in sec',fontsize=14)
plt.title('t_comp_avg for XTC 100x and 300x on Stampede',fontsize=14)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1)
XTCmeans100.plot(x='Block',y='t_comp_max',ax=axes,label='XTC100',style='-*')
XTCmeans300.plot(x='Block',y='t_comp_max',ax=axes,label='XTC300',style='-o')
axes.set_xticks([16,32,48,64])
axes.set_xlabel('Blocks',fontsize=14)
axes.set_ylabel('Time in sec',fontsize=14)
plt.title('t_comp_max for XTC 100x and 300x on Stampede',fontsize=14)
```
