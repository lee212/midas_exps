```python
singleCorePython1 = pd.DataFrame.from_csv('stragglers/tidy_data/SingleCore_104675.csv')
singleCoreCpp1 = pd.DataFrame.from_csv('stragglers/tidy_data/timings_104675.csv')

singleCorePython2 = pd.DataFrame.from_csv('stragglers/tidy_data/SingleCore_52338.csv')
singleCoreCpp2 = pd.DataFrame.from_csv('stragglers/tidy_data/timings_52338.csv')

singleCorePython3 = pd.DataFrame.from_csv('stragglers/tidy_data/SingleCore_34892.csv')
singleCoreCpp3 = pd.DataFrame.from_csv('stragglers/tidy_data/timings_34892.csv')
```

```python
histSingleCorePython,edgesSingleCorePython = np.histogram(singleCorePython1['Compute'].values,bins=100)
histSingleCoreCpp,edgesSingleCoreCpp = np.histogram((singleCoreCpp1/1000000.0)['Execute'].values,bins=100)

histSingleCorePython1 = histSingleCorePython/len(singleCorePython1['Compute'].values)
widthSingleCorePython1 = 0.9 * (edgesSingleCorePython[1] - edgesSingleCorePython[0])
centerSingleCorePython1 = (edgesSingleCorePython[:-1] + edgesSingleCorePython[1:]) / 2

histSingleCoreCpp1 = histSingleCoreCpp/len(singleCoreCpp1['Execute'].values)
widthSingleCoreCpp1 = 0.9 * (edgesSingleCoreCpp[1] - edgesSingleCoreCpp[0])
centerSingleCoreCpp1 = (edgesSingleCoreCpp[:-1] + edgesSingleCoreCpp[1:]) / 2
```
