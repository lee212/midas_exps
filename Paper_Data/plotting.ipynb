```python
def kmeans_exp_reader(filename):
    import csv
    exp_name = filename.split('.')[0]
    
    with open(filename,'rb') as open_file:
        data = csv.reader(open_file,delimiter=',')
    
        exp_data=dict()
        for row in data:
            exp_data[row[1]] = row[2:]
        
    mapper_iter1 = list()
    mapper_iter2 = list()
    reducer_iter1 = list()
    reducer_iter2 = list()
    
    detailed_data = {'FirstMapperStarted1': float(exp_data['Mapper-0-1'][3]),
                     'LastMapperEnded1'   : float(exp_data['Mapper-0-1'][5]),
                     'ReducerStarted1'    : float(exp_data['Reducer-0'][3]),
                     'ReducerEnded1'      : float(exp_data['Reducer-0'][5]),
                     'FirstMapperStarted2': float(exp_data['Mapper-1-1'][3]),
                     'LastMapperEnded2'   : float(exp_data['Mapper-1-1'][5]),
                     'ReducerStarted2'    : float(exp_data['Reducer-1'][3]),
                     'ReducerEnded2'      : float(exp_data['Reducer-1'][5])
                    }
    
    for key,value in exp_data.iteritems():
        if key.startswith('Mapper-0'):
            mapper_iter1.append(float(value[4]) - float(value[3]))
            
            if value[3] < detailed_data['FirstMapperStarted1']:
                detailed_data['FirstMapperStarted1'] = float(value[3])
            
            if value[5] < detailed_data['LastMapperEnded1']:
                detailed_data['LastMapperEnded1'] = float(value[5])
        elif key.startswith('Mapper-1'):
            mapper_iter2.append(float(value[4]) - float(value[3]))
            
            
            if value[3] < detailed_data['FirstMapperStarted2']:
                detailed_data['FirstMapperStarted2'] = float(value[3])
            
            if value[5] < detailed_data['LastMapperEnded2']:
                detailed_data['LastMapperEnded2'] = float(value[5])
                
        elif key.startswith('Reducer-0'):
            reducer_iter1.append(float(value[4]) - float(value[3]))
        elif key.startswith('Reducer-1'):
            reducer_iter2.append(float(value[4]) - float(value[3]))
            
    return exp_name,detailed_data,mapper_iter1,reducer_iter1,mapper_iter2,reducer_iter2
```
