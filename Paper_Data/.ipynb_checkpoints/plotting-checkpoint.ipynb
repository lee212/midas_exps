```python
pilot_file1=open('Stampede/vanilla_json/rp.session.RadicalExp.iparask.016822.0041.json')
pilot_file2=open('Stampede/vanilla_json/rp.session.RadicalExp.iparask.016822.0042.json')
pilot_file3=open('Stampede/vanilla_json/rp.session.RadicalExp.iparask.016822.0043.json')
pilot1=json.load(pilot_file1)
pilot2=json.load(pilot_file2)
pilot3=json.load(pilot_file3)
pilot_file1.close()
pilot_file2.close()
pilot_file3.close()
units_startup=list()
i=0
for elements in pilot1['unit']:
    for state in elements['statehistory']:
        if state['state']=='ExecutingPending':
            temp1=state['timestamp']
        elif state['state']=='Executing':
            temp2=state['timestamp']
    units_startup.append(temp2-temp1)

for elements in pilot2['unit']:
    for state in elements['statehistory']:
        if state['state']=='ExecutingPending':
            temp1=state['timestamp']
        elif state['state']=='Executing':
            temp2=state['timestamp']
    units_startup.append(temp2-temp1)

for elements in pilot3['unit']:
    for state in elements['statehistory']:
        if state['state']=='ExecutingPending':
            temp1=state['timestamp']
        elif state['state']=='Executing':
            temp2=state['timestamp']
    units_startup.append(temp2-temp1)
print units_startup.__len__()
mean_units_startup=np.mean(units_startup)
std_units_startup=np.std(units_startup)
```
