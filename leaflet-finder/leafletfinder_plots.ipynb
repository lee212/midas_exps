```python
stampede_frame1_128c_1809w_1_128 = open('Iteration3/reports/stampede_frame1_128c_1809w_1-128.csv')
stampede_frame1_128c_1809w_2_128 = open('Iteration3/reports/stampede_frame1_128c_1809w_2-128.csv')
stampede_frame1_128c_1809w_3_128 = open('Iteration3/reports/stampede_frame1_128c_1809w_3-128.csv')
stampede_frame1_16c_1809w_1_16 = open('Iteration3/reports/stampede_frame1_16c_1809w_1-16.csv')
stampede_frame1_16c_1809w_2_16 = open('Iteration3/reports/stampede_frame1_16c_1809w_2-16.csv')
stampede_frame1_16c_1809w_3_16 = open('Iteration3/reports/stampede_frame1_16c_1809w_3-16.csv')
stampede_frame1_32c_1809w_1_32 = open('Iteration3/reports/stampede_frame1_32c_1809w_1-32.csv')
stampede_frame1_32c_1809w_2_32 = open('Iteration3/reports/stampede_frame1_32c_1809w_2-32.csv')
stampede_frame1_32c_1809w_3_32 = open('Iteration3/reports/stampede_frame1_32c_1809w_3-32.csv')
stampede_frame1_64c_1809w_1_64 = open('Iteration3/reports/stampede_frame1_64c_1809w_1-64.csv')
stampede_frame1_64c_1809w_2_64 = open('Iteration3/reports/stampede_frame1_64c_1809w_2-64.csv')
stampede_frame1_64c_1809w_3_64 = open('Iteration3/reports/stampede_frame1_64c_1809w_3-64.csv')



leaflet_128c_1809w_1_128 = csv.reader(stampede_frame1_128c_1809w_1_128)
leaflet_128c_1809w_2_128 = csv.reader(stampede_frame1_128c_1809w_2_128)
leaflet_128c_1809w_3_128 = csv.reader(stampede_frame1_128c_1809w_3_128)
leaflet_16c_1809w_1_16 = csv.reader(stampede_frame1_16c_1809w_1_16)
leaflet_16c_1809w_2_16 = csv.reader(stampede_frame1_16c_1809w_2_16)
leaflet_16c_1809w_3_16 = csv.reader(stampede_frame1_16c_1809w_3_16)
leaflet_32c_1809w_1_32 = csv.reader(stampede_frame1_32c_1809w_1_32)
leaflet_32c_1809w_2_32 = csv.reader(stampede_frame1_32c_1809w_2_32)
leaflet_32c_1809w_3_32 = csv.reader(stampede_frame1_32c_1809w_3_32)
leaflet_64c_1809w_1_64 = csv.reader(stampede_frame1_64c_1809w_1_64)
leaflet_64c_1809w_2_64 = csv.reader(stampede_frame1_64c_1809w_2_64)
leaflet_64c_1809w_3_64 = csv.reader(stampede_frame1_64c_1809w_3_64)
```

```python
leaflet_128_1_dist = list()
for row in leaflet_128c_1809w_1_128:
    if row[1].startswith('Euclidean'):
        leaflet_128_1_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_128_1_con = float(row[-1])
        
leaflet_128_2_dist = list()
for row in leaflet_128c_1809w_2_128:
    if row[1].startswith('Euclidean'):
        leaflet_128_2_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_128_2_con = float(row[-1])

leaflet_128_3_dist = list()
for row in leaflet_128c_1809w_3_128:
    if row[1].startswith('Euclidean'):
        leaflet_128_3_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_128_3_con = float(row[-1])
        
leaflet_64_1_dist = list()
for row in leaflet_64c_1809w_1_64:
    if row[1].startswith('Euclidean'):
        leaflet_64_1_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_64_1_con = float(row[-1])

leaflet_64_2_dist = list()
for row in leaflet_64c_1809w_2_64:
    if row[1].startswith('Euclidean'):
        leaflet_64_2_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_64_2_con = float(row[-1])

leaflet_64_3_dist = list()
for row in leaflet_64c_1809w_3_64:
    if row[1].startswith('Euclidean'):
        leaflet_64_3_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_64_3_con = float(row[-1])


leaflet_32_1_dist = list()
for row in leaflet_32c_1809w_1_32:
    if row[1].startswith('Euclidean'):
        leaflet_32_1_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_32_1_con = float(row[-1])
        
leaflet_32_2_dist = list()
for row in leaflet_32c_1809w_2_32:
    if row[1].startswith('Euclidean'):
        leaflet_32_2_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_32_2_con = float(row[-1])

leaflet_32_3_dist = list()
for row in leaflet_32c_1809w_3_32:
    if row[1].startswith('Euclidean'):
        leaflet_32_3_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_32_3_con = float(row[-1])

leaflet_16_1_dist = list()
for row in leaflet_16c_1809w_1_16:
    if row[1].startswith('Euclidean'):
        leaflet_16_1_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_16_1_con = float(row[-1])
        
leaflet_16_2_dist = list()
for row in leaflet_16c_1809w_2_16:
    if row[1].startswith('Euclidean'):
        leaflet_16_2_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_16_2_con = float(row[-1])

leaflet_16_3_dist = list()
for row in leaflet_16c_1809w_3_16:
    if row[1].startswith('Euclidean'):
        leaflet_16_3_dist.append(float(row[-1]))
    elif row[1].startswith('ConnComp'):
        leaflet_16_3_con = float(row[-1])
```

```python
leaflet_128_dist = np.mean([np.max(leaflet_128_1_dist),np.max(leaflet_128_2_dist),np.max(leaflet_128_3_dist)])
leaflet_128_con = np.mean([leaflet_128_1_con,leaflet_128_2_con,leaflet_128_3_con])
leaflet_64_dist = np.mean([np.max(leaflet_64_1_dist),np.max(leaflet_64_2_dist),np.max(leaflet_64_3_dist)])
leaflet_64_con = np.mean([leaflet_64_1_con,leaflet_64_2_con,leaflet_64_3_con])
leaflet_32_dist = np.mean([np.max(leaflet_32_1_dist),np.max(leaflet_32_2_dist),np.max(leaflet_32_3_dist)])
leaflet_32_con = np.mean([leaflet_32_1_con,leaflet_32_2_con,leaflet_32_3_con])
leaflet_16_dist = np.mean([np.max(leaflet_16_1_dist),np.max(leaflet_16_2_dist),np.max(leaflet_16_3_dist)])
leaflet_16_con = np.mean([leaflet_16_1_con,leaflet_16_2_con,leaflet_16_3_con])
```
